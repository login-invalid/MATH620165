```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
from sklearn.datasets import load_files
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import re

X, y = [], []
for i in range(1,7):
    emails = load_files(f"./input/enron{i}")
    X = np.append(X, emails.data)
    y = np.append(y, emails.target)

classes = emails.target_names
print(f"X.shape: {X.shape:}")
print(f"y.shape: {y.shape}")
print("\n")
#print(f"Example X[0]: {X[0]}")
#print("\n")
print(f"Classe X[0]: {y[0]} ({classes[int(y[0])]})")
from sklearn.model_selection import train_test_split
X_filtered = []

for email in range(0, len(X)):
    text = re.sub(r'\\r\\n', ' ', str(X[email]))
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    text = re.sub(r'^b\s+', '', text)
    text = text.lower()
    X_filtered.append(text)

#print(f"Example X[0]: {X_filtered[0]}")
print("\n")
print(f"Classes X[0]: {y[0]} ({classes[int(y[0])]})")

X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, test_size=0.3)
clf_pipeline = Pipeline([
    ('tfidf_vectorizer', TfidfVectorizer()),
    ('classifier', MultinomialNB())])

clf_pipeline.fit(X_train, y_train)
predictions = clf_pipeline.predict(X_train)
score = accuracy_score(y_train,predictions)
print(f"Accuracy on training set: {score*100:.2f}%")

predictions_test = clf_pipeline.predict(X_test)
score = accuracy_score(y_test, predictions_test)
print(f"Accuracy on testing set: {score*100:.2f}%")

print("Test set confusion matrix:")
y_true = pd.Series(y_test, name='Real')
y_pred = pd.Series(predictions_test, name=' Prediction')
pd.crosstab(y_true, y_pred)
```
